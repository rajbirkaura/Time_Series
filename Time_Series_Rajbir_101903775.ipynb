```python
# Importing all required libraries

import numpy as np
import pandas as pd
from math import sqrt
import matplotlib.pyplot as plt
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_squared_error
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# Creating model, training and predicting on test data

def train_time_series_with_folds(temp_df, section, horizon=1):
    x = temp_df.drop(columns=['Para-9', 'Para-10',	'Para-11',	'Para-12',	'Para-13'], axis=1)
    y = temp_df['Para-9']

  # Take 10th year of dataset for validation
    x_train, x_test = x.iloc[:-horizon, :], x.iloc[-horizon:, :]
    y_train, y_test = y.iloc[:-horizon], y.iloc[-horizon:]

    model = LGBMRegressor(random_state=42)
    model.fit(x_train, y_train)
    predictions = model.predict(x_test)
    rmse = np.round(sqrt(mean_squared_error(y_test, predictions)), 3)

    actual = y_test.values
    rmse_df = pd.DataFrame(columns=['Para-9', 'Para-10',	'Para-11',	'Para-12',	'Para-13'])
    rmse_df.loc[0, 'Para-9'] = actual[0]
    rmse_df.loc[1, 'Para-9'] = np.round(predictions[0], 3)
    rmse_df.loc[2, 'Para-9'] = rmse
    rmse_df.to_csv("/content/RMSE_comparisons/sec"+str(section)+".csv", index=False)
```

```python
# Function for making graphs for comparing actual and predicted values

import numpy as np
import matplotlib.pyplot as plt

list1 = []

def comparison_graph(rmse_df, section):
    mean_rmse = np.round(rmse_df.iloc[2, :].mean(axis=0), 3)
    list1.append(mean_rmse)
    line_1 = rmse_df.iloc[0, :]
    line_2 = rmse_df.iloc[1, :]
    fig, ax = plt.subplots()
    ax.plot(line_1, color = 'green', label = 'Actual')
    ax.plot(line_2, color = 'red', label = 'Predicted')
    plt.title("RMSE: "+str(mean_rmse))
    ax.legend()
    plt.savefig("/content/Plots/sec"+str(section)+".png")
    plt.close()
```

```python
# Making graphs for all sections

for i in range(1, 102):
    if (i == 92):
        continue
    rmse_df = pd.read_csv("/content/RMSE_comparisons/sec"+str(i)+".csv")
    comparison_graph(rmse_df, i)
```

```python
# Calculating final RMSE of the model

f_rmse = sum(list1)/len(list1)
print("RMSE: ", np.round(f_rmse, 3))
```
